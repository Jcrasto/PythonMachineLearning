```python
import pandas
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
import xgboost as xgb
```

```python
fires = pandas.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/forest-fires/forestfires.csv')
fires.head()
```

```python
fires.describe()
```

```python
fires.corr()
```

```python
#X = fires.loc[:, fires.columns != 'area']
#X = fires[fires.columns.difference(['area', 'month','day'])]
X = fires.loc[:,"temp":"rain"] #repository notes said these cols were only ones used for best estimates 
y = fires['area']
train_X, test_X, train_y, test_y = train_test_split(X, y,test_size=0.3)
```

```python
dtree = DecisionTreeRegressor()
dtree.fit(train_X,train_y)
dtree.score(test_X,test_y)
```

```python
rfor = RandomForestRegressor()
rfor.fit(train_X,train_y)
rfor.score(test_X,test_y)
```

```python
ada = AdaBoostRegressor()
ada.fit(train_X,train_y)
ada.score(test_X,test_y)
```

```python
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(train_X,train_y)
knn.score(test_X,test_y)
```

```python
xg_class = xgb.XGBRegressor()
xg_class.fit(train_X,train_y)
xg_class.score(test_X,test_y)
```
