```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

%matplotlib inline
```

```python
col_names = ['Alcohol','Malic acid','Ash','Alcalinity_of_ash','Magnesium','Total_phenols','Flavanoids','Nonflavanoid_phenols','Proanthocyanins','Color_intensity','Hue','OD280/OD315_of_diluted_wines','Proline']

wine = pandas.read_csv(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data",
    header=None,
    names=col_names,
    index_col=False,
)
```

```python
wine.corr()
```

```python
X_train ,X_test = train_test_split(wine,test_size=0.3)
Kmean = KMeans(n_clusters=7)
Kmean.fit(X_train)
Kmean.cluster_centers_

y_train = Kmean.labels_
y_test = Kmean.predict(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```
