```python
import pandas
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsRegressor
```

```python
col_names = ['Alcohol','Malic acid','Ash','Alcalinity_of_ash','Magnesium','Total_phenols','Flavanoids','Nonflavanoid_phenols','Proanthocyanins','Color_intensity','Hue','OD280/OD315_of_diluted_wines','Proline']

wine = pandas.read_csv(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data",
    header=None,
    names=col_names,
    index_col=False,
)
```

```python
X = wine.loc[:, wine.columns != 'Alcohol']
y = wine['Alcohol']
train_X, test_X, train_y, test_y = train_test_split(X, y)
```

```python
neighbors = range(1,10)
maxk = 0
maxk_score = 0
for k in neighbors:
    #print(str(k) + " Regressor")
    #knn = KNeighborsRegressor(n_neighbors=k)
    #knn.fit(train_X,train_y)
    #print(str((knn.score(test_X,test_y) + knn.score(X,y))/2) + '    ' + str(knn.score(test_X,test_y) - knn.score(X,y)))
    
    #Classfication models fit better
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_X,train_y)
    if knn.score(test_X,test_y) > maxk_score:
        maxk_score = knn.score(test_X,test_y)
        maxk = k
print(str(maxk) + ": " + str(maxk_score))
```
