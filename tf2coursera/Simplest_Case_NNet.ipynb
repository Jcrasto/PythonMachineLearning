```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import math
```

### Input shape 1

```python
x = np.array(range(1,10)).astype('double')
```

```python
lin_y = 3 * x + 2
exp_y = np.exp(x)
log_y = np.log(x)
```

```python
lin_model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
lin_model.compile(optimizer='sgd', loss='mean_squared_error')
lin_model.fit(x, lin_y, epochs=50)

exp_model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
exp_model.compile(optimizer='sgd', loss='logcosh')
exp_model.fit(x, exp_y, epochs=50)

log_model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
log_model.compile(optimizer='sgd', loss='mean_squared_error')
log_model.fit(x, log_y, epochs=50)
```

```python
check = 11
print(lin_model.predict([check]))
print(exp_model.predict([check]))
print(log_model.predict([check]))
```

```python
print(3 * check + 2)
print(math.exp(check))
print(math.log(check))
```

```python
print(lin_model.predict([check]) - (3 * check + 2))
print(exp_model.predict([check]) - math.exp(check))
print(log_model.predict([check]) - math.log(check))
```

### Input Shape 2

```python
x = np.array(range(10)).astype('double')
y = np.array(range(-5,5)).astype('double')
z = x * y
x = np.column_stack((x,y))
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=(2,))])
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(x, z, epochs=50)
```

```python
check = np.array([[ 3, 8]])
model.predict(check)
```
